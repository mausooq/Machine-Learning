```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('canada_per_capita_income.csv')
df.tail(5)
```

```python
df.columns
```

```python
plt.xlabel("year")
plt.ylabel("Per Capita Income (US$)")
plt.scatter(df['year'], df['perCapitaIncome'], color='red', marker='*')
```

```python
reg  = linear_model.LinearRegression()
reg.fit(df[['year']],df['perCapitaIncome'])
```

```python
reg.predict([[2020]])
```

```python
plt.xlabel('Year')
plt.ylabel('Per Capita Income (US$)')
plt.scatter(df['year'], df['perCapitaIncome'], color='red', label='Actual Data')

plt.plot(df['year'], reg.predict(df[['year']]), color='blue', label='Regression Line')
```

```python
%matplotlib inline
plt.xlabel('Year')
plt.ylabel('Per Capita Income (US$)')
plt.scatter(df['year'], df['perCapitaIncome'], color='red', label='Actual Data')

plt.plot(df['year'], reg.predict(df[['year']]), color='blue', label='Regression Line')
plt.scatter(2020, reg.predict([[2020]]), color='green', s=100, label='Prediction for 2020', edgecolor='black')

plt.legend()
```

```python
%matplotlib inline
plt.xlabel('Year')
plt.ylabel('Per Capita Income (US$)')
plt.scatter(df['year'], df['perCapitaIncome'], color='red', label='Actual Data')

plt.plot(df['year'], reg.predict(df[['year']]), color='blue', label='Regression Line')
plt.scatter(2020, reg.predict([[2020]]), color='green', s=100, label='Prediction for 2020', edgecolor='black')
plt.axvline(x=2020, color='purple', linestyle='--', label='Year 2020')
plt.axhline(y= reg.predict([[2020]]), color='orange', linestyle='--', label='Predicted Income')

plt.legend()
```
