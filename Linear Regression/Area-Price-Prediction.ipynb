```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df=pd.read_csv("Housing.csv")
df
```

```python
%matplotlib inline
plt.xlabel('area (sqr ft)')
plt.ylabel('US $')
plt.scatter(df.area,df.price,color='green',marker='*')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']],df.price)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
135.78767123*3300+180616.43835616432
```

```python
%matplotlib inline
plt.xlabel('area (sqr ft)')
plt.ylabel('US $')
plt.scatter(df.area,df.price,color='red')
plt.plot(df.area,reg.predict(df[['area']]),color='blue')
plt.scatter(2020, reg.predict([[2020]]), color='green', s=100, label='Prediction for 2020', edgecolor='black')
plt.legend()
```

```python
dff = pd.read_csv('Area.csv')
dff
```

```python
p = reg.predict(dff)
```

```python
dff['price'] =p
```

```python
dff.to_csv("Area-Price.csv",index =False)
```
